## 1. Importing Necessary Libraries 

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

## 2. Import Data

```python
from sklearn import datasets
breast_cancer_data=datasets.load_breast_cancer()
breast_cancer_data
```

```python
breast_cancer_data.DESCR
```

```python
breast_cancer_data.data
```

**Malignant Cancer is More Dangarous**, we need to reduce false negetive

```python
cancer_data_df=pd.DataFrame(data=breast_cancer_data.data, columns=breast_cancer_data.feature_names)
cancer_data_df['target']=breast_cancer_data.target
cancer_data_df
```

## 3. Data Understanding

```python
cancer_data_df.shape
```

```python
cancer_data_df.isna().sum()
```

```python
cancer_data_df.dtypes
```

## 4. Model Building

```python
X=cancer_data_df.drop(labels='target', axis=1)
y=cancer_data_df[['target']]
```

```python
X.shape, y.shape
```

```python
y.value_counts()
```

```python
plt.pie(x=y.value_counts(), labels=['Benign','Malignant'],autopct='%1.0f%%')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.20, random_state=12, shuffle=True, stratify=y)
#Stratify - preserves the same proportions of examples in each class as observed in the original dataset.
```

```python
X_train.shape, y_train.shape
```

```python
plt.pie(x=y_train.value_counts(), labels=['Benign','Malignant'],autopct='%1.0f%%')
plt.show()
```

```python
X_test.shape, y_test.shape
```

```python
plt.pie(x=y_test.value_counts(), labels=['Benign','Malignant'],autopct='%1.0f%%')
plt.show()
```

## 5. Model Training

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier(criterion='gini', max_depth=3, class_weight={0:2,1:1})
dt_model.fit(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression(class_weight={0:2,1:1})
log_model.fit(X_train, y_train)
```

## 6. Model Testing || 7. Model Evaluation

### Train Data

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_pred_train=dt_model.predict(X_train)
```

```python
y_pred_train_log=log_model.predict(X_train)
```

```python
print(confusion_matrix(y_train, y_pred_train))
```

```python
print(confusion_matrix(y_train, y_pred_train_log))
```

```python
print(classification_report(y_train, y_pred_train))
```

```python
print(classification_report(y_train, y_pred_train_log))
```

### Test Data

```python
y_pred_test=dt_model.predict(X_test)
```

```python
y_pred_test_log=log_model.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred_test))
```

```python
print(confusion_matrix(y_test, y_pred_test_log))
```

```python
print(classification_report(y_test, y_pred_test))
```

```python
print(classification_report(y_test, y_pred_test_log))
```
